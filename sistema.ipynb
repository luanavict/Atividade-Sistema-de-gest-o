```python



while True:
    print("\n=== MENU PRINCIPAL ===")
    print("1 - Sistema de Cadastro de Alunos")
    print("2 - Controle de Despesas Pessoais")
    print("3 - Cardápio da Lanchonete")
    print("4 - Controle de Estoque da Lanchonete")
    print("0 - Sair")

    opcao_principal = input("Escolha: ")

    if opcao_principal == "1":
        alunos = []
        while True:
            print("\n=== SISTEMA DE CADASTRO DE ALUNOS ===")
            print("1 - Cadastrar aluno")
            print("2 - Listar alunos")
            print("3 - Buscar aluno pelo nome")
            print("4 - Remover aluno")
            print("5 - Verificar se aluno existe")
            print("6 - Mostrar maior e menor nome")
            print("7 - Transformar nomes em MAIÚSCULAS")
            print("8 - Remover último aluno")
            print("0 - Voltar ao menu principal")

            opc = input("Escolha: ")

            if opc == "1":
                nome = input("Nome do aluno: ").strip()
                alunos.append(nome)
                print("Aluno cadastrado com sucesso!")

            elif opc == "2":
                if alunos:
                    print("Alunos cadastrados:")
                    for a in alunos:
                        print("-", a)
                else:
                    print("Nenhum aluno cadastrado.")

            elif opc == "3":
                nome = input("Digite o nome para buscar: ")
                if nome in alunos:
                    print("Aluno encontrado!")
                else:
                    print("Aluno não encontrado.")

            elif opc == "4":
                nome = input("Nome do aluno a remover: ")
                if nome in alunos:
                    alunos.remove(nome)
                    print("Aluno removido!")
                else:
                    print("Aluno não encontrado.")

            elif opc == "5":
                nome = input("Digite o nome a verificar: ")
                print("Existe na lista?", nome in alunos)

            elif opc == "6":
                if alunos:
                    maior = max(alunos)
                    menor = min(alunos)
                    print("Maior (ordem alfabética):", maior)
                    print("Menor (ordem alfabética):", menor)
                else:
                    print("Nenhum aluno cadastrado.")

            elif opc == "7":
                alunos = [a.upper() for a in alunos]
                print("Todos os nomes convertidos para maiúsculas!")

            elif opc == "8":
                if alunos:
                    removido = alunos.pop()
                    print("Último aluno removido:", removido)
                else:
                    print("Lista vazia!")

            elif opc == "0":
                break
            else:
                print("Opção inválida!")



    elif opcao_principal == "2":
        despesas = []
        print("\n=== CONTROLE DE DESPESAS ===")
        print("Digite valores de despesas (ou 0 para parar)")

        while True:
            valor = float(input("Despesa: "))
            if valor == 0:
                break
            despesas.append(valor)

        if len(despesas) == 0:
            print("Nenhuma despesa registrada.")
        else:
            total = sum(despesas)
            media = total / len(despesas)
            maior = max(despesas)
            menor = min(despesas)

            print(f"\nTotal gasto: R$ {total:.2f}")
            print(f"Média de gastos: R$ {media:.2f}")
            print(f"Maior gasto: R$ {maior:.2f}")
            print(f"Menor gasto: R$ {menor:.2f}")


    elif opcao_principal == "3":
        itens = ["Coxinha", "Pastel", "Suco", "Refrigerante", "Hambúrguer"]
        precos = [6.00, 7.00, 5.00, 6.50, 12.00]
        total = 0

        while True:
            print("\n=== CARDÁPIO ===")
            for i in range(len(itens)):
                print(f"{i + 1} - {itens[i]} (R$ {precos[i]:.2f})")
            print("0 - Encerrar pedido")

            opc = int(input("Escolha o número do item: "))
            if opc == 0:
                break
            elif 1 <= opc <= len(itens):
                total += precos[opc - 1]
                print(f"{itens[opc - 1]} adicionado ao pedido!")
            else:
                print("Opção inválida!")

        print(f"\nTotal do pedido: R$ {total:.2f}")


    elif opcao_principal == "4":
        produtos = []
        quantidades = []

        while True:
            print("\n=== CONTROLE DE ESTOQUE ===")
            print("1 - Adicionar produto")
            print("2 - Atualizar quantidade")
            print("3 - Listar produtos")
            print("4 - Mostrar produtos com estoque zerado")
            print("0 - Voltar ao menu principal")
            opc = input("Escolha: ")

            if opc == "1":
                nome = input("Nome do produto: ")
                qtd = int(input("Quantidade: "))
                produtos.append(nome)
                quantidades.append(qtd)
                print("Produto adicionado!")

            elif opc == "2":
                nome = input("Nome do produto: ")
                if nome in produtos:
                    indice = produtos.index(nome)
                    nova_qtd = int(input("Nova quantidade: "))
                    quantidades[indice] = nova_qtd
                    print("Quantidade atualizada!")
                else:
                    print("Produto não encontrado.")

            elif opc == "3":
                if len(produtos) > 0:
                    print("\n--- Lista de produtos ---")
                    for i in range(len(produtos)):
                        print(f"{produtos[i]}: {quantidades[i]} unidades")
                else:
                    print("Nenhum produto cadastrado.")

            elif opc == "4":
                print("\nProdutos com estoque zerado:")
                for i in range(len(produtos)):
                    if quantidades[i] == 0:
                        print("-", produtos[i])

            elif opc == "0":
                break
            else:
                print("Opção inválida!")


    elif opcao_principal == "0":
        print("Saindo do sistema... Até logo!")
        break

    else:
        print("Opção inválida! Tente novamente.")
```
